```python
import pandas as pd
import json
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

denest pli2en_ncped.json

```python
dictionaryList = json.load(open("pli2en_ncped.json"))
y = [{}]
for entry in dictionaryList:
    a = entry.get("homonyms")
    if type(a) is list:
        for k in a:
            k_def = k.get("definition")
            if type(k_def) is list:
                for l in k_def:
                    y.append({
                        "entry": entry.get("entry"),
                        "grammar": entry.get("grammar"),
                        "definition": l,
                        "xr": entry.get("xr")
                    })
            else:
                y.append({
                    "entry": entry.get("entry"),
                    "grammar": k.get("grammar"),
                    "definition": k_def,
                    "xr": k.get("xr")
                })
    else:
        k_def = entry.get("definition")
        #print(k_def)
        if type(k_def) is list:
            for l in k_def:
                y.append({
                    "entry": entry.get("entry"),
                    "grammar": entry.get("grammar"),
                    "definition": l,
                    "xr": entry.get("xr")
                })
        else:
            y.append({
                "entry": entry.get("entry"),
                "grammar": entry.get("grammar"),
                "definition": entry.get("definition"),
                "xr": entry.get("xr")
            })

json.dump(y, open("pli2en_ncped_denested.json", "w+"), ensure_ascii=False)

df = pd.read_json("pli2en_ncped_denested.json")
df.to_csv("pli2en_ncped.csv", sep="\t")
```

cleanup pli2en_ncped.json

```python
sc = pd.read_csv("pli2en_ncped.csv", sep="\t", index_col=0)
sc.columns = ["pali", "grammar", "meaning", "from"]
sc.dropna(subset=["pali"], axis=0, inplace=True)
sc.dropna(subset=["meaning"], axis=0, inplace=True)

sc["pali"] = sc["pali"].str.replace("ṁ", "ṃ")
sc["pali"] = sc["pali"].str.replace("a\(t\)$", "anta")
sc["pali"] = sc["pali"].str.replace("at\)$", "anta")
sc["pali"] = sc["pali"].str.replace("a\(r\)$", "ar")
sc["pali"] = sc["pali"].str.replace("ar\)", "ar")
sc["pali"] = sc["pali"].str.replace("i\(n\)$", "ī")
sc["pali"] = sc["pali"].str.replace("ī\(n\)", "ī")
sc["pali"] = sc["pali"].str.replace("ī\(n", "ī")
sc["pali"] = sc["pali"].str.replace("in$", "ī")
sc["pali"] = sc["pali"].str.replace("\(k\)", "k")
sc["pali"] = sc["pali"].str.replace("\(r\)$", "")
sc["pali"] = sc["pali"].str.replace("\(ṃ\)$", "")
sc["pali"] = sc["pali"].str.replace("\(n\)$", "")
sc["pali"] = sc["pali"].str.replace("\(s\)$", "s")
sc["pali"] = sc["pali"].str.replace("\(g\)", "g")
sc["pali"] = sc["pali"].str.replace("\(d\)", "d")
sc["pali"] = sc["pali"].str.replace("\(n\)", "n")
sc["pali"] = sc["pali"].str.replace("\(b\)", "b")
sc["pali"] = sc["pali"].str.replace("\(m\)", "m")
sc["pali"] = sc["pali"].str.replace("\(p\)", "p")
sc["pali"] = sc["pali"].str.replace("\(s\)", "s")
sc["pali"] = sc["pali"].str.replace("\(t\)", "t")
sc["pali"] = sc["pali"].str.replace(" \(va\)", "va")
sc["pali"] = sc["pali"].str.replace("\-", "")
sc["pali"] = sc["pali"].str.replace("\'", "")
sc["pali"] = sc["pali"].str.replace("\’", "")
sc["pali"] = sc["pali"].str.replace(",.*$", "")

sc.to_csv("pli2en_ncped_cleaned.csv", sep="\t", index=False)
sc
```

reduce & cleanup dpd

```python
dpd = pd.read_csv("/home/bhikkhu/Bodhirasa/Dropbox/dpd/csvs/dpd.csv", sep="\t")
# print(dpd.columns)

dpd = dpd[["Pāli1", "Grammar", "Meaning IN CONTEXT", "Derived from"]]

dpd.columns = ["pali", "grammar", "meaning", "from"]

dpd["pali"] = dpd["pali"].str.replace(fr" \d*$", "")

dpd
```

merge and generate differences

```python
df3  = pd.merge(dpd, sc, on=["pali"], how='right', indicator='exists')
df3  = df3.loc[df3['exists'] == 'right_only']
df3.dropna(axis=1, inplace=True, how='all')

df3["meaning_y"].dropna()
df3.to_csv("differences.csv", sep="\t", index=None)
df3
```
